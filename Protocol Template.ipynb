## Checklist for required features and best practices

### Must-haves
* Links to additional material
* Fully running example
* Strict reproducible instructions for example case
* Describe parametrization to deal with other data sets + not that reproducible instructions for additional use cases
* Strictly define limitations, therein minimal system requirements (RAM, disk space, CPU power to run protocol (docker) OK time)
* Shared/generic interface to the different protocols: proteomics-academy.org webpage? Showing overview of protocols +  their workflow visualizations would be nice.
* Protocol/settings versioning (part of docker image versions) 
* With possibility to use specific version of the protocol/settings if needed using sort of a “client” to the protocol -> will be there when providing dockers as biocontainers
* Control for unsuitable data and error messages, well-defined access to log-files from the protocol runs
* Provide log-files in specific folders and also give feedback when running the commands in the example or other cases
* Validation of intermediate results QC (validation but also as documentation, do the results make sense?)

### Best practice
* Providing links of each/most software tools to bio.tools
* Workflow diagram/scheme, maybe with controlled vocabulary (e.g. dot file instructions in * Jupyter notebook, software requirement analysis document RAD-like format)
* Well-defined descriptions of how and where to change parameters and input files of the protocol pipeline (e.g fasta file)
* Via web interface change settings file
* Change in scripts
* Should work on all platforms (probably obsolete as most of them run on Docker)
* Presentation of the intermediate results (graphs, overview of the tables) and possibility to discontinue program execution
* Using it in batch mode on example case
* Use case diagram
* Optimally restricted user input for parameter settings to be more robust on the long run
* Encourage verbosity
* I/O formats, it can be zipped, hdf or other. Allow for customization. ( ultimately user controls where the organisation of the data) 
* On the long run: input file handler?
* Optional features (O)
* Video tutorials
* Easily distributed computation
