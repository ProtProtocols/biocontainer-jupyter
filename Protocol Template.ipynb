## Checklist for required features and best practices

### Must-haves

#### General
* Links to additional material
* Fully running example
* Strictly defined limitations, therein minimal system requirements (RAM, disk space, CPU power to run protocol on docker image)
* Protocol versioning (part of docker image versions) 
* Well-defined access to log-files from the protocol runs via _LOG_ folder
* All input files including the ones for the example are in folder _IN_
* All output files and intermediate results are available in folder _OUT_

#### Example

* Strict reproducible instructions for example case
* Describe parametrization to extend use case to other input data or different analyses
* Control for incorrect intermediate results and error messages

### Best practice
#### General
* Provide links of used software tools in bio.tools
* Workflow diagram/scheme using controlled vocabulary (e.g. dot file instructions in Jupyter notebook named by EDAM ontology terms)
* Separate Jupyter notebooks for example use case and generalized workflow that allows adjustment to different input data formats and different use cases
* Allow customization of I/O formats (e.g. zipped files).  

#### Example

* Well-defined descriptions of how and where to change parameters and input files of the protocol pipeline (e.g fasta file)
* Interactive setting of parameters in notebook (e.g. selection in list or ruler for numerical input with defined limits)
* Visual evaluation of the intermediate results (figures, overview of the tables) and possibility to discontinue program execution
* One-click analysis of entire example use case
* Verbosity (e.g. automatic output of log-files into notebook cells
